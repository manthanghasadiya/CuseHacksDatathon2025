```python
import pandas as pd
import numpy as np

file_path = '/content/drive/MyDrive/Datathon/cityline.csv'

try:
  df = pd.read_csv(file_path)
  print(df.head()) # Display the first few rows of the DataFrame
except FileNotFoundError:
  print(f"Error: File not found at {file_path}. Please check the file path.")
except pd.errors.ParserError:
    print(f"Error: Could not parse the CSV file at {file_path}.  Please check the file format.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
df
```

```python
df.drop_duplicates(inplace=True)
df.dropna(axis=1, how='all', inplace=True)
df.fillna('Null', inplace=True)

df
```

```python
df = df.dropna(subset=['Lat', 'Lng'])
df
```

```python
df_a = df[df['Acknowledged_at_local'] != "Null"]
df_final = df_a[df_a['Closed_at_local'] != "Null"]
```

```python
df_final
```

```python
df['Acknowledged_at_local'] = df['Acknowledged_at_local'].replace('Null', np.nan)
df['Closed_at_local'] = df['Closed_at_local'].replace('Null', np.nan)

df['Created_at_local'] = pd.to_datetime(df['Created_at_local'], errors='coerce')
df['Acknowledged_at_local'] = pd.to_datetime(df['Acknowledged_at_local'], errors='coerce')
df['Closed_at_local'] = pd.to_datetime(df['Closed_at_local'], errors='coerce')
```

```python
df_aa = df[df['Acknowledged_at_local'].notna() & df['Closed_at_local'].isna()]
df_ac = df[df['Acknowledged_at_local'].isna() & df['Closed_at_local'].notna()]
```

```python
df_ac
```

```python
# Data Insight A: Requests Acknowledged but Not Closed

in_progress_counts = df_aa.groupby('Agency_Name').size().sort_values(ascending=False)
print("Number of Acknowledged-But-Unclosed Requests by Agency:")
print(in_progress_counts)

now = pd.Timestamp.now()
df_aa['time_since_ack'] = (now - df_aa['Acknowledged_at_local']).dt.total_seconds() / 3600  # hours

agency_stuck_time = df_aa.groupby('Agency_Name')['time_since_ack'].mean().sort_values(ascending=False)
print("\nAverage Hours Since Acknowledgment (Unclosed) by Agency:")
print(agency_stuck_time)
```

```python
# Data Insight B: Requests Closed but Never Acknowledged

stealth_closes = df_ac.groupby('Agency_Name').size().sort_values(ascending=False)
print("Number of Requests Closed Without Acknowledgment by Agency:")
print(stealth_closes)

df_ac['Minutes_to_Close'] = pd.to_numeric(df_ac['Minutes_to_Close'], errors='coerce')
avg_close_no_ack = df_ac.groupby('Agency_Name')['Minutes_to_Close'].mean().sort_values(ascending=False)
print("\nAverage Minutes to Close (No-Ack Requests) by Agency:")
print(avg_close_no_ack)
```

```python
# Sort df_aa by how long it's been open since acknowledgment
df_aa['time_since_ack'] = (pd.Timestamp.now() - df_aa['Acknowledged_at_local']).dt.total_seconds() / 3600
longest_open = df_aa.sort_values('time_since_ack', ascending=False).head(10)
print("Top 10 Longest Open Requests (Hours Since Acknowledgement):")
print(longest_open[['Id','Agency_Name','Created_at_local','Acknowledged_at_local','time_since_ack']])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

features = df_pivot.drop(['Week'], axis=1)
X = StandardScaler().fit_transform(features)

dbscan = DBSCAN(eps=2, min_samples=5)
labels = dbscan.fit_predict(X)
df_pivot['ClusterLabel'] = labels
df_pivot
```

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

df_rc = df_final.copy()
df_rc['Created_at_local'] = pd.to_datetime(df['Created_at_local'], errors='coerce')

df_rc['Week'] = df_rc['Created_at_local'].dt.to_period('W').dt.start_time
df_grouped = df_rc.groupby(['Week']).size().reset_index(name='RequestCount')

X = df_grouped[['RequestCount']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

iso = IsolationForest(contamination=0.01, random_state=42)
df_grouped['OutlierLabel'] = iso.fit_predict(X_scaled)  # 1 = normal, -1 = outlier

df_anomalies = df_grouped[df_grouped['OutlierLabel'] == -1]
print(df_anomalies)
```

```python
df['Created_at_local'] = pd.to_datetime(df['Created_at_local'], errors='coerce')

df['Month'] = df['Created_at_local'].dt.to_period('M').dt.start_time

all_months = df['Month'].unique()
all_agencies = [
    'Sanitation', 'Garbage, Recycling & Graffiti', 'Streets & Transportation',
    'Housing & Property Maintenance', 'Sanitation Carts', 'Water & Sewage',
    'Green Spaces, Trees & Public Utilities', 'Animals',
    'Streets, Sidewalks & Transportation', 'Parking & Vehicles',
    'Feedback to the City', 'Health, Safety & Social Services'
]

all_combinations = pd.MultiIndex.from_product([all_months, all_agencies], names=['Month', 'Agency_Name']).to_frame(index=False)

grouped = df.groupby(['Month', 'Agency_Name', 'Lat', 'Lng']).size().reset_index(name='Count')

complete_grouped = pd.merge(all_combinations, grouped, on=['Month', 'Agency_Name'], how='left').fillna({
    'Lat': 43.05,
    'Lng': -76.14,
    'Count': 0
})
complete_grouped.sort_values(by='Month', inplace=True)

complete_grouped
```

```python
all_agencies = [
    'Sanitation', 'Garbage, Recycling & Graffiti', 'Streets & Transportation',
    'Housing & Property Maintenance', 'Sanitation Carts', 'Water & Sewage',
    'Green Spaces, Trees & Public Utilities', 'Animals',
    'Streets, Sidewalks & Transportation', 'Parking & Vehicles',
    'Feedback to the City', 'Health, Safety & Social Services'
]

color_discrete_map = {
    'Sanitation': '#636EFA',  # Blue
    'Garbage, Recycling & Graffiti': '#EF553B',  # Red-Orange
    'Streets & Transportation': '#00CC96',  # Green-Teal
    'Housing & Property Maintenance': '#AB63FA',  # Purple
    'Sanitation Carts': '#FFA15A',  # Orange
    'Water & Sewage': '#19D3F3',  # Light Blue
    'Green Spaces, Trees & Public Utilities': '#FF6692',  # Pink
    'Animals': '#B6E880',  # Light Green
    'Streets, Sidewalks & Transportation': '#FF97FF',  # Light Magenta
    'Parking & Vehicles': '#FECB52',  # Mustard
    'Feedback to the City': '#FFA2C6',  # Pinkish
    'Health, Safety & Social Services': '#AA9674'  # Brownish
}

fig = px.scatter_mapbox(
    complete_grouped,
    lat='Lat',
    lon='Lng',
    color='Agency_Name',
    size='Count',
    animation_frame='Month',
    category_orders={'Agency_Name': all_agencies},
    color_discrete_map=color_discrete_map,
    mapbox_style='carto-positron',
    zoom=10,
    center={"lat": 43.05, "lon": -76.14},
    title="All Agencies - Complaints Over Time (Custom Colors)"
)

fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})
fig.show()
```
