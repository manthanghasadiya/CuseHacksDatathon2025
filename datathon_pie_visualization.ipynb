```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
# Convert date column to datetime
df_ac['Created_at_local'] = pd.to_datetime(df_ac['Created_at_local'], errors='coerce')

# Drop rows with no valid date if needed
df_ac.dropna(subset=['Created_at_local'], inplace=True)
# Count total requests by Agency + Category
grouped_counts = (
    df_ac.groupby(['Agency_Name', 'Category'])
    .size()
    .reset_index(name='Total_Requests')
)

# For each agency, pick the top 5 categories
top5_cats = (
    grouped_counts.groupby('Agency_Name', group_keys=True)
    .apply(lambda x: x.nlargest(5, 'Total_Requests'))
    .reset_index(drop=True)
)

# This DataFrame now has only the top 5 categories for each agency.
```

```python
# Get the agencies that appear in the top5_cats table
unique_agencies = top5_cats['Agency_Name'].unique()

for agency in unique_agencies:
    # Identify this agency’s top 5 categories
    sub_top5 = top5_cats[top5_cats['Agency_Name'] == agency]['Category'].unique()

    # Filter the main DataFrame for this agency
    df_agency = df_ac[df_ac['Agency_Name'] == agency].copy()

    # Create a single figure to plot all top 5 categories
    plt.figure(figsize=(12, 6))

    for cat in sub_top5:
        # Filter for this category
        df_cat = df_agency[df_agency['Category'] == cat].copy()

        # Group by month
        df_monthly = (
            df_cat.groupby(pd.Grouper(key='Created_at_local', freq='M'))
            .size()
            .reset_index(name='requests')
        )

        # If insufficient data (e.g., fewer than 2 data points), skip
        if len(df_monthly) < 2:
            continue

        # Rename columns for Prophet
        df_monthly.rename(columns={'Created_at_local': 'ds', 'requests': 'y'}, inplace=True)

        # Initialize Prophet
        model = Prophet(
            yearly_seasonality=True,
            weekly_seasonality=False,
            daily_seasonality=False
        )
        model.fit(df_monthly)

        # Forecast next 12 months
        future = model.make_future_dataframe(periods=12, freq='M')
        forecast = model.predict(future)

        # Plot forecast line
        plt.plot(forecast['ds'], forecast['yhat'], label=cat)

        # Plot confidence interval
        plt.fill_between(
            forecast['ds'],
            forecast['yhat_lower'],
            forecast['yhat_upper'],
            alpha=0.2
        )

    # Force y-axis to start at 0
    plt.ylim(bottom=0)

    plt.title(f"Monthly Forecast for Top 5 Categories - {agency}")
    plt.xlabel("Date")
    plt.ylabel("Number of Requests")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```
